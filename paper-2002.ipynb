```python
import matplotlib.pyplot as plt
import numpy as np

# Initial and new positions
initial_distance = 176.74  # d
new_distance = initial_distance - 120  # d - 120

# Angles in degrees
angle1 = 19
angle2 = 47

# Building height
height = 60.88

# Convert angles to radians
angle1_rad = np.deg2rad(angle1)
angle2_rad = np.deg2rad(angle2)

# Creating figure
fig, ax = plt.subplots()

# Plotting the building
plt.plot([0, 0], [0, height], 'k-', lw=2)

# Plotting the initial triangle
plt.plot([initial_distance, 0], [0, height], 'b-', label=f'Initial angle {angle1}°')
plt.plot([initial_distance, initial_distance], [0, height], 'b--')

# Plotting the new triangle
plt.plot([new_distance, 0], [0, height], 'r-', label=f'New angle {angle2}°')
plt.plot([new_distance, new_distance], [0, height], 'r--')

# Adding points for clarity
plt.plot(0, 0, 'bo')
plt.plot(0, new_distance, 'ro')
plt.plot(initial_distance, height, 'go')

# Annotations
plt.text(initial_distance / 2, height / 2, 'd', fontsize=12, color='blue')
plt.text(new_distance / 2, height / 2, 'd - 120', fontsize=12, color='red')
plt.text(-5, height / 2, 'h', fontsize=12, color='green')

# Angles annotations
plt.text(initial_distance - 10, 5, f'{angle1}°', fontsize=12, color='blue', ha='right')
plt.text(new_distance - 10, 5, f'{angle2}°', fontsize=12, color='red', ha='right')

# Plot the angles in the triangles
plt.plot([initial_distance, initial_distance - 10*np.cos(angle1_rad)], 
         [0, 10*np.sin(angle1_rad)], 'b-')
plt.plot([new_distance, new_distance - 10*np.cos(angle2_rad)], 
         [0, 10*np.sin(angle2_rad)], 'r-')

# Axes properties
plt.xlim(-20, initial_distance + 20)
plt.ylim(0, height + 20)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

# Labels
plt.xlabel('Distance (m)')
plt.ylabel('Height (m)')

# Title and legend
plt.title('Angle of Elevation of a Building')
plt.legend()

# Show plot
plt.show()
```
